## Latihan

1. Import Data:
 * Buka dataset penjualan_online.csv, lalu tampilkan 5 baris pertama.


```python
import pandas as pd
df = pd.read_csv("penjualan_online.csv")
df.head(5)
```

2. Summary Tables:
* Tampilkan informasi dataset menggunakan info() dan describe().
* Hitung total pendapatan per kategori produk dengan groupby().


```python
print("Menampilkan informasi menggunakan info()")
print(df.info())
print('\n')
print("Menampilkan informasi menggunakan describe()")
print(df.describe())
print('\n')
print("Menampilkan total pendapatan per kategori produk")
df.groupby('Kategori')['Pendapatan'].sum()
```

3. Visualisasi Data:
* Buat histogram untuk harga produk.
* Buat boxplot untuk melihat outlier harga.
* Buat bar chart jumlah penjualan per kategori.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat histogram untuk harga produk
plt.figure()
plt.hist(df['Harga'], bins=8)
plt.xlabel('Harga')
plt.ylabel('Frekuensi')
plt.title('Histogram Harga Produk')
plt.show()
```

```python
# Membuat boxplot untuk melihat outlier harga
plt.figure()
plt.boxplot(df['Harga'])
plt.ylabel('Harga')
plt.title('Boxplot Harga Produk')
plt.show()
```

Terdapat 2 data outlier yang cukup jauh

```python
# Membuat bar chart jumlah penjualan per kategori

category_counts = df['Kategori'].value_counts()
plt.figure()
print(category_counts)
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Kategori')
plt.ylabel('Jumlah Penjualan')
plt.title('Jumlah Penjualan per Kategori')
```

Barang dengan penjualan terbanyak berasal dari kategori Elektronik

**Membuat scatter plot antara Harga dan Quantity**

```python
plt.figure()
plt.scatter(df["Harga"], df["Quantity"])
plt.xlabel("Harga")
plt.ylabel("Quantity")
plt.title("Hubungan harga dan Quantity")
plt.show()

print("\n")

sns.regplot(x="Harga", y="Quantity", data=df)
plt.title("Scatter Plot dengan Garis Regresi")
plt.xlabel("Harga")
plt.ylabel("Quantity")
plt.show()
```

Hasil nya garis lurus yang berarti tidak memiliki korelasi antar 2 varibel tersebut
